# Explore reference and pipeline outputs

Side-by-side comparison of summer orthophoto, synthetic winter output, and intermediate snow/terrain layers for manual QA.

```python
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
import rasterio

PROJECT_ROOT = Path("..").resolve()
TILE_ID = "davos_001"
intermediate = PROJECT_ROOT / "data/intermediate/tiles" / TILE_ID
output = PROJECT_ROOT / "data/output/tiles" / TILE_ID


def read_rgb(path: Path) -> np.ndarray:
    with rasterio.open(path) as src:
        data = src.read()
    if data.shape[0] >= 3:
        return np.moveaxis(data[:3], 0, -1)
    return data[0]


def read_band(path: Path, band: int = 1) -> np.ndarray:
    with rasterio.open(path) as src:
        return src.read(band)


summer = read_rgb(intermediate / "rgb_summer.tif")
winter = read_rgb(output / f"{TILE_ID}_winter_rgb.tif")
snow_fraction = read_band(output / f"{TILE_ID}_snow_fraction.tif")
hillshade = read_band(intermediate / "terrain_features.tif", band=6)

fig, axes = plt.subplots(2, 2, figsize=(12, 12))
axes[0, 0].imshow(summer)
axes[0, 0].set_title("Summer RGB")
axes[0, 1].imshow(winter)
axes[0, 1].set_title("Winter RGB")
axes[1, 0].imshow(snow_fraction, cmap="Blues", vmin=0, vmax=1)
axes[1, 0].set_title("Snow fraction")
axes[1, 1].imshow(hillshade, cmap="gray")
axes[1, 1].set_title("Winter hillshade")
for ax in axes.ravel():
    ax.axis("off")
plt.tight_layout()
plt.show()
```
